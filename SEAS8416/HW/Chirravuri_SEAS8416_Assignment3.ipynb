## Question - 1

A cybersecurity firm implements a new intrusion detection system (IDS) designed to reduce the average response time to detect and mitigate cyberattacks. Historically, the average response time with the old system was 15 minutes. After deploying the new IDS, the firm collects a sample of 50 incident response times.

1. Formulate the null hypothesis (H₀) and the alternative hypothesis (Hₐ) to test whether the new IDS reduces the average response time.

2. Once you have your hypotheses, propose a statistical test you would use to compare response times and explain why it is appropriate for this scenario.

3. Research one tail and two tailed tests and identify which is appropriate.


## Answer for Question - 1

### Hypothesis Formulation

1. **Null Hypothesis (H₀):**  
   The new IDS does not reduce the average response time, or it is the same as the old system.  
   **H₀: μ ≥ 15 minutes**  
   where **μ** is the mean response time with the new IDS.  
   Example: If the average response time with the new IDS is 15 minutes or longer, the null hypothesis holds true, indicating no improvement.

2. **Alternative Hypothesis (Hₐ):**  
   The new IDS reduces the average response time compared to the old system.  
   **Hₐ: μ < 15 minutes**  
   Example: If the average response time with the new IDS is found to be 12 minutes, it supports the alternative hypothesis, showing the system is faster.

---

### Statistical Test

To compare the response times, a **one-sample t-test** is appropriate because:

1. **Known historical mean:**  
   The old system’s response time is 15 minutes, which acts as the baseline for comparison.

2. **Reasonable sample size:**  
   The firm collected a sample of 50 response times, which is large enough to assume that the sampling distribution of the mean is approximately normal due to the Central Limit Theorem.

3. **Unknown population standard deviation:**  
   Since the population standard deviation of the response times is not provided, the t-test is more appropriate than a z-test.

---

### One-Tailed vs. Two-Tailed Test

1. **One-Tailed Test:**  
   Since the primary goal is to test if the new IDS **reduces** the response time, a one-tailed test is suitable. This focuses only on detecting whether the new system is faster.  
   - **Example:** Suppose the sample mean is 13 minutes, with statistical evidence supporting μ < 15. This would lead to rejecting the null hypothesis.

2. **Two-Tailed Test:**  
   A two-tailed test would be used if the goal were to detect **any difference** (either reduction or increase) in response time. However, the scenario focuses specifically on reduction.  
   - **Example:** A two-tailed test would be relevant if the firm wanted to test whether the new system's average response time differed from 15 minutes in either direction.

---

### Practical Scenarios for Clarity

- **One-Tailed Test Scenario:**  
  A cybersecurity firm wants to ensure the new IDS reduces response time to address time-sensitive attacks, such as ransomware spreading across a network. A one-tailed test will focus on improvements, confirming the system’s efficiency.  

- **Two-Tailed Test Scenario:**  
  The firm may want to evaluate whether the system performs differently (better or worse) across various network configurations. In this case, the two-tailed test evaluates performance variations.

---

### Conclusion

A **one-tailed t-test** is most appropriate to test whether the new IDS significantly reduces response time compared to the old system. The focus is on determining if the new system provides faster incident response times, which is critical for mitigating cyberattacks effectively.

## Question - 4

Use a statistical test in Python to determine if the data for `chol` (cholesterol levels) is normally distributed.

## Requirements:
1. Identify the **p-value** and **test statistic** from the results.
2. Explain the following in your own words:
   - **Null Hypothesis (H₀):** What does it represent in this context?
   - **Alternative Hypothesis (H₁):** What does it represent in this context?
3. Provide an interpretation of the p-value and test statistic in the context of normality.

## Answer for Question - 4

To test whether the data for chol is normally distributed, we can use the **Shapiro-Wilk** test. This test is commonly used and appropriate for small to moderate sample sizes.

Here's the code to perform the Shapiro-Wilk test:

````python
import pandas as pd  # Import pandas for data manipulation
from scipy.stats import shapiro  # Import the Shapiro-Wilk test from scipy.stats

# Load the dataset
data = pd.read_csv('heart.csv')  # Read the CSV file into a pandas DataFrame

# Extract the 'chol' column
chol_data = data['chol']  # Select the 'chol' column which contains cholesterol data

# Perform the Shapiro-Wilk test
stat, p_value = shapiro(chol_data)  # Compute the test statistic and p-value for normality

# Print the test statistic and p-value
print(f"Shapiro-Wilk Test Statistic: {stat}")  # Display the test statistic (W)
print(f"P-value: {p_value}")  # Display the p-value indicating the significance of the test
````

## Results of Shapiro-Wilk Normality Test for `chol`

### Hypotheses for the Shapiro-Wilk Test

- **Null Hypothesis (H₀):** The data for `chol` follows a normal distribution.
- **Alternative Hypothesis (H₁):** The data for `chol` does not follow a normal distribution.

---

### Test Results

- **Shapiro-Wilk Test Statistic:** 0.9468815274790511  
- **P-value:** 5.364847008526812e-09  

---

## Interpreting the Results

### P-value:
- **If p-value > 0.05:** Fail to reject the null hypothesis. This suggests the data is consistent with a normal distribution.
- **If p-value ≤ 0.05:** Reject the null hypothesis. This indicates the data significantly deviates from a normal distribution.

### Test Statistic:
- The test statistic (W) measures how well the data matches the expected normal distribution.
- Values closer to 1 indicate closer alignment with normality.

---

### Explanation

- Since the p-value (5.36e-09) is much smaller than 0.05, we reject the null hypothesis. This indicates that the `chol` data is not normally distributed.

---

### Summary

The Shapiro-Wilk test determines whether a dataset resembles a normal distribution.

- A very small p-value (e.g., ≤ 0.05) suggests that the data significantly deviates from normality.
- In this case, the p-value is extremely small, so we conclude that the `chol` data does **not** follow a normal distribution.

   - Assume the data is a **sample** of the population.

````python
# Answer to Question - 5
# Author: Viswanath S Chirravuri
# Date: November 24th, 2024

import pandas as pd  # Import pandas library for handling data

# Load the dataset
data = pd.read_csv('heart.csv')

# My custom function to calculate the standard deviation
def vis_standard_deviation_function(data_list):
    """
    Calculate the standard deviation of a dataset (sample).
    Parameters:
        data_list (list): List of numerical data.
    Returns:
        float: Standard deviation of the dataset.
    """
    try:
        # Check if the input is a list
        if not isinstance(data_list, list):  # Ensure the data is passed as a list
            raise TypeError("Input data must be a list.")  # Raise an error if it's not a list

        # Check if the list is empty
        if len(data_list) == 0:  # Check if the list is empty
            raise ValueError("Input list is empty.")  # Raise an error if the list is empty

        # Calculate mean
        n = len(data_list)  # Get the number of elements in the list
        mean = sum(data_list) / n  # Calculate the mean (average) of the data

        # Calculate squared deviations from the mean
        squared_deviations = [(x - mean) ** 2 for x in data_list]  # Compute squared deviations for each element

        # Calculate variance (for a sample)
        variance = sum(squared_deviations) / (n - 1)  # Divide by (n-1) for sample variance

        # Return standard deviation (which is square root of variance)
        return variance ** 0.5  # Return the square root of the variance as the standard deviation
    except Exception as e:
        print(f"Error: {e}")  # Print the error message if an exception is raised
        return None  # Return None if there's an error

# Test the function with a feature from the heart dataset
chol_data = data['chol'].tolist()  # Convert 'chol' column to a list for the custom function
custom_std = vis_standard_deviation_function(chol_data)  # Calculate standard deviation using the custom function

# Calculate standard deviation using pandas for validation
built_in_std = data['chol'].std()  # Use pandas' built-in method to calculate standard deviation

# Print the results
print(f"Custom Standard Deviation (chol): {custom_std}")  # Print the custom function's result
print(f"Built-in Standard Deviation (chol): {built_in_std}")  # Print the built-in function's result

# Compare the results
if abs(custom_std - built_in_std) < 1e-6:  # Allow a small tolerance for floating-point comparison
    print("The custom function matches the built-in library result.")  # If results are close, print success
else:
    print("The custom function does not match the built-in library result.")  # If results differ, print failure
````

## Question - 7

Find and display the **Jaccard distance** between the following sets:

```python
set1 = {"apple", "banana", "cherry", "date", "fig"}
set2 = {"banana", "cherry", "grape", "kiwi", "melon"}
```

**Note:** You can use `scipy`, but only load the method you need. Using sets directly will not work for some `scipy` methods, so you may need to convert a set to another structure (e.g., binary vectors) before applying the method.

````python
# Answer to Question - 7
# Author: Viswanath S Chirravuri
# Date: November 23rd, 2024

# Import the required method for computing Jaccard distance
from scipy.spatial.distance import jaccard

# Define the sets
set1 = {"apple", "banana", "cherry", "date", "fig"}
set2 = {"banana", "cherry", "grape", "kiwi", "melon"}

# Combine all unique elements from both sets
# This is done by taking the union of both sets
all_elements = list(set1.union(set2))

# Convert each set into a binary vector representation
# The binary vector indicates presence (1) or absence (0) of elements from 'all_elements'
vector1 = [1 if elem in set1 else 0 for elem in all_elements]  # Binary vector for set1
vector2 = [1 if elem in set2 else 0 for elem in all_elements]  # Binary vector for set2

# Compute the Jaccard distance between the two binary vectors
# Jaccard distance is calculated as the ratio of the difference to the union of the two sets
distance = jaccard(vector1, vector2)

# Print the Jaccard distance
print("Jaccard Distance:", distance)
````

## Answer to Question-8
# Simplification Using De Morgan's Laws

We aim to simplify the expression:

**((A ∩ B) ∪ C)^c**

---

## Step 1: Apply De Morgan's Law to the Outer Negation

De Morgan's laws are:
1. **(X ∪ Y)^c = X^c ∩ Y^c**
2. **(X ∩ Y)^c = X^c ∪ Y^c**

The given expression is **((A ∩ B) ∪ C)^c**, which involves a complement applied to a union. Using the first De Morgan's law, distribute the complement over the union:

**((A ∩ B) ∪ C)^c = (A ∩ B)^c ∩ C^c**

This breaks down the complement of the union into the intersection of the complements of the individual components: **(A ∩ B)** and **C**.

---

## Step 2: Simplify (A ∩ B)^c Using De Morgan's Law

The term **(A ∩ B)^c** is a complement applied to an intersection. Using the second De Morgan's law, distribute the complement over the intersection:

**(A ∩ B)^c = A^c ∪ B^c**

This means the complement of **(A ∩ B)** is equivalent to the union of **A^c** and **B^c**.

---

## Step 3: Substitute Back Into the Expression

Substitute **(A ∩ B)^c = A^c ∪ B^c** into the equation from Step 1:

**((A ∩ B) ∪ C)^c = (A^c ∪ B^c) ∩ C^c**

---

## Final Simplified Expression

The final simplified result is:

**(A^c ∪ B^c) ∩ C^c**

---

# Explanation of the Result

- **(A^c ∪ B^c):** This represents the condition where either **A** is false (**A^c**) or **B** is false (**B^c**).
- **∩ C^c:** This adds the requirement that **C** is also false (**C^c**).
