### TASK-1

A medical test is designed to detect a certain disease. The test is not perfect, and we know the following:

- The probability that a person has the disease is 5%
- The probability that the test correctly identifies a person with the disease is 98%.

**Question**: What is the probability that a person has the disease given that the test result is positive?  Calculate and provide the answer as code.

```python
# Answer to Task-1
# Author: Viswanath S Chirravuri
# Date: December 9th 2024

# Given probabilities
P_D = 0.05  # Probability of having the disease
P_not_D = 1 - P_D  # Probability of not having the disease i.e., 1 - 0.05 = 0.95
P_T_given_D = 0.98  # Probability test is positive given disease
P_T_given_not_D = 1 - P_T_given_D  # Probability test is positive given no disease i.e., 1 - 0.98 = 0.02

# Compute the total probability of a positive test result
P_T = P_T_given_D * P_D + P_T_given_not_D * P_not_D

# Compute the probability of having the disease given a positive test result
P_D_given_T = (P_T_given_D * P_D) / P_T

# Print the result by limiting the value to 2 digits after the dot.
print(f"The probability that a person has the disease given a positive test result is: {P_D_given_T*100:.2f}%")
```

### TASK-2

Using the credit card data set provided on Blackboard.  Perform anomaly detection via the following steps.
- Perform EDA (use visuals where appropriate).  Pay attention to the class balance.  Is the data balanced?
- Test train split the data.  Ensure class balance is maintained.
- Scale the data with a standard scaler.
    - Add a markdown cell and discuss the problem of information leakage from train to test if scaling takes place before test train split.
- Use a grid search to arrive at best hyperparameters for 2 models.
    - Isolation Forest
    - DBSCAN or OCSVM (OCSVM may take a while)
    - Display the optimal parameters
- Train and test models.
    - Provide performance metrics. (e.g., F1, Precision, Recall)
    - Show confusion matrices and ROC AUC curves.
- Use PCA to reduce features to account for 80% of the variance in the data.
    - Train and test again to compare.
- Analyze results.  What conclusions can you make about the models’ ability to detect anomalies.  Which approaches worked best?

```python
# Answer to Task-2
# Author: Viswanath S Chirravuri
# Date: December 9th 2024

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score

# Load the dataset
data_path = "creditcard.csv"
df = pd.read_csv(data_path)

# Exploratory Data Analysis (EDA)
print("Dataset Overview:")
print(df.info())
print("Class Distribution:")
print(df['Class'].value_counts())

# Visualize class balance
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Class', palette='Set2', legend=False, hue='Class')
plt.title("Class Distribution")
plt.show()

# Split data into features and target
X = df.drop(columns=['Class'])
y = df['Class']

# Train-test split (maintaining class balance)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Scale the data (avoid information leakage)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Discussing information leakage
"""
Scaling before train-test split can lead to information leakage, as the scaling parameters (mean and standard deviation)
are calculated using the entire dataset. This could result in biased models during training and testing. To avoid this,
fit the scaler only on the training set and apply it to the test set separately.
"""

# Grid search for hyperparameters
# Isolation Forest
iso_forest_params = {
    'n_estimators': [50, 100, 150],
    'max_samples': ['auto', 0.5, 0.75],
    'contamination': [0.01, 0.05, 0.1],
    'random_state': [42]
}
iso_forest = IsolationForest()
grid_iso = GridSearchCV(iso_forest, iso_forest_params, scoring='accuracy', cv=3, n_jobs=-1)
grid_iso.fit(X_train_scaled)

# Optimal parameters for Isolation Forest
print("Best parameters for Isolation Forest:")
print(grid_iso.best_params_)

# DBSCAN or One-Class SVM (e.g., OCSVM)
ocsvm_params = {
    'kernel': ['rbf', 'linear'],
    'nu': [0.01, 0.05, 0.1],
    'gamma': ['scale', 'auto']
}
ocsvm = OneClassSVM()
grid_ocsvm = GridSearchCV(ocsvm, ocsvm_params, scoring='accuracy', cv=3, n_jobs=-1)
grid_ocsvm.fit(X_train_scaled)

# Optimal parameters for OCSVM
print("Best parameters for OCSVM:")
print(grid_ocsvm.best_params_)

# Train and test models
iso_forest_best = grid_iso.best_estimator_
ocsvm_best = grid_ocsvm.best_estimator_

# Predictions
iso_preds = iso_forest_best.predict(X_test_scaled)
ocsvm_preds = ocsvm_best.predict(X_test_scaled)

# Converting predictions to binary for consistency
iso_preds = np.where(iso_preds == 1, 0, 1)
ocsvm_preds = np.where(ocsvm_preds == 1, 0, 1)

# Performance Metrics
def evaluate_model(y_true, y_pred):
    print("Classification Report:")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))

    fpr, tpr, _ = roc_curve(y_true, y_pred)
    auc_score = roc_auc_score(y_true, y_pred)
    print(f"AUC Score: {auc_score}")

    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, label=f"AUC = {auc_score:.2f}")
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.title("ROC Curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

# Evaluate models
print("Isolation Forest Results:")
evaluate_model(y_test, iso_preds)

print("OCSVM Results:")
evaluate_model(y_test, ocsvm_preds)

# PCA for dimensionality reduction
pca = PCA(n_components=0.8)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Retrain and test models with PCA
iso_forest_best.fit(X_train_pca)
ocsvm_best.fit(X_train_pca)

iso_pca_preds = iso_forest_best.predict(X_test_pca)
ocsvm_pca_preds = ocsvm_best.predict(X_test_pca)

# Convert PCA predictions to binary
iso_pca_preds = np.where(iso_pca_preds == 1, 0, 1)
ocsvm_pca_preds = np.where(ocsvm_pca_preds == 1, 0, 1)

# Evaluate models with PCA
print("Isolation Forest Results with PCA:")
evaluate_model(y_test, iso_pca_preds)

print("OCSVM Results with PCA:")
evaluate_model(y_test, ocsvm_pca_preds)
```
